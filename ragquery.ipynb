```python

from sentence_transformers import SentenceTransformer
import chromadb
from pypdf import PdfReader
import requests
from io import BytesIO

pdf_file="/content/FLAP-SAM.pdf"
#we get the pdf through reader
pdf_reader =PdfReader(pdf_file)
documents = [page.extract_text() for page in pdf_reader.pages]
#extract pages
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(documents)

client = chromadb.Client()
collection = client.create_collection("ragdocs1")
```

```python

for i, (doc, emb) in enumerate(zip(documents, embeddings)):
    collection.add(documents=[doc], embeddings=[emb], ids=[str(i)])

def rag_query(question, k=3):
    query_embedding = model.encode([question])[0]
    results = collection.query(query_embeddings=[query_embedding], n_results=k)
    context = " ".join(results['documents'][0])
    return f"Context: {context[:500]}...\n\nQuestion: {question}"
rag_query("amount of  communication cost reduction achieved")
```
